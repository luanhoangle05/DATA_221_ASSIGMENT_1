```python
import math
def circle_area_coverage(radiusofcircle1,radiusofcircle2):
    if radiusofcircle1<=0 or radiusofcircle2<=0:
        print("Your input are not correct, they have to be postive integers")
    else:
        areaofcircle1=math.pi*(radiusofcircle1**2)
        areaofcircle2=math.pi*(radiusofcircle2**2)
        largerarea=max(areaofcircle1,areaofcircle2)
        smallerarea=min(areaofcircle1,areaofcircle2)
        coveragepercentage=(smallerarea/largerarea)*100
        return coveragepercentage
print(circle_area_coverage(3,4))
```
