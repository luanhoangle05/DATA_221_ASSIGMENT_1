```python
def distribution(my_list_of_numbers):
    my_result={}
    my_sorted_list=sorted(my_list_of_numbers)
    total=len(my_sorted_list)
    for value in my_sorted_list:
        if value not in my_result:
            count=0
            for number in my_sorted_list:
                if number<=value:
                    count+=1
            percentage=(count/total)*100
            my_result[value]={percentage}
    return my_result
print(distribution([3,1,2,3,4,2]))

```
