```python
def nested_dictionary_from_string(string_list):
    result_dict={}
    for word in string_list:
        length_of_string=len(word)
        if length_of_string%2==0:
            parity="even"
        else:
            parity="odd"
        result_dict[word]={"length":length_of_string, "parity":parity}
    return result_dict
print(nested_dictionary_from_string(["data","science"]))
```
