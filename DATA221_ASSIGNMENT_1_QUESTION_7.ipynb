```python
def convert_into_time(seconds_input):
    if seconds_input<0 or seconds_input>=86400:
        return "Invalid input, seconds must be between 0 and 86399"
    else:
        hour_in_24=seconds_input//3600
        minutes=(seconds_input%3600)//60
        second=seconds_input%60
        if hour_in_24 < 12:
            period="AM"
        else:
            period="PM"
        hour_in_12=hour_in_24%12
        if hour_in_12==0:
            hour_in_12=12
        return (f"The time is {hour_in_12}:{minutes:02d}:{second:02d} {period}")


print(convert_into_time(86399))


```
