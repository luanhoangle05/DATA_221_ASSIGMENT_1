```python
def compute_power(base_value, exponent_value):
    return base_value**exponent_value
pairs=[[5,2],[3,-1],[4,3],[2,0]]
validResult=[]
for base_value, exponentValue in pairs: # unpack items in smaller lists
    if exponentValue<0:
        continue
    else:
        validResult.append(compute_power(base_value,exponentValue))
print(validResult)
```
