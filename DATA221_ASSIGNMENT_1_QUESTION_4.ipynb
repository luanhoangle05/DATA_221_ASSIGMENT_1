```python
from random import random
values=[random() for i in range (0,20)]
x=random()
values=sorted(values)
print("sorted list is: ",values)
print("X is: ",x)
found_index=None
for i in range (0,len(values)):
    if values[i]>=x:
        found_index=i
        break
if found_index is not None:
    print("The first matching index is: ",found_index)
else:
    print("There is no value greater than or equal to x")

```
